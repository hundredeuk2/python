# 2. 도형 그리기

```python
# 256 X 256크기의 사이즈 페이지에 3채널(RGB 색상)을 사용한 도화지를 만듦

img = np.zeros((256,256,3), np.uint8) 
```

```python
# 직선 그리기
cv2.line(img, (0,0), (256,256), (255,0,0), 3)
cv2.line(img, (256,0), (0,256), (0,255,0), 10)

plt.imshow(img)
cv2.waitKey(0) 
cv2.destroyAllWindows()

# 아래의 그림을 보았을 때, 뜻하는 파라미터들은
# cv2.line(img, 시작점, 끝점, (R, G, B), 두께) 다음과 같다고 볼 수 있다.
```

```python
img = np.zeros((256,256,3), np.uint8) 
```

```python
#직사각형: cv2.retengle(img, 1번점, 4번점, 색(B,G,R), 두께) 
cv2.rectangle(img, (200, 200), (256, 256), (0, 255, 0),5) 
#원: cv2.circle(img, 중심, 반지름, 색, 두께) 
cv2.circle(img, (100,63), 63, (0, 255, 255), -1) 
#글자 넣기: cv2.putText(img, 넣을 텍스트, 시작 점, 폰트, 폰트 크기, 색, 글자 두꼐) 
cv2.putText(img, '글 자 넣 기', (10, 100), cv2.FONT_HERSHEY_SIMPLEX, 4, (255, 255, 255), 4) 

#화면 크기 조정 가능 
plt.imshow(img)
cv2.waitKey(0) 
cv2.destroyAllWindows()


# 아래의 그림으로 보아 한글은 글자넣기가 잘 안되는 것으로 파악
# 어떻게 폰트 적용을 하게 되면 가능할지도(?) 방법 모색 가능
```
